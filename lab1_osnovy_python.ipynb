#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text):
    vowels = 'aeiou'
    count = 0
    
    text_lower = text.lower()
    
    for char in text_lower:
        if char in vowels:
            count += 1
            
    return count
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n < 0:
        raise ValueError("Число должно быть положительным")
    
    if n < 10:
        return 0
    
    steps = 0
    current = n
    
    while current >= 10:
        product = 1
        for digit in str(current):
            product *= int(digit)
        
        current = product
        steps += 1
    
    return steps
```

```python
class TestMagic:
    
    @staticmethod
    def test_single_digit():
        result = magic(4)
        expected = 0
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_single_digit пройден")
    
    @staticmethod
    def test_example_39():
        result = magic(39)
        expected = 3
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_example_39 пройден")
    
    @staticmethod
    def test_example_999():
        result = magic(999)
        expected = 4
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_example_999 пройден")
    
    @staticmethod
    def test_two_digits():
        result = magic(25)
        expected = 2
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_two_digits пройден")
    
    @staticmethod
    def test_three_digits():
        result = magic(123)
        expected = 1
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_three_digits пройден")
    
    @staticmethod
    def test_with_zero():
        result = magic(10)
        expected = 1
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_with_zero пройден")
    
    @staticmethod
    def test_another_example():
        result = magic(68)
        expected = 3
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_another_example пройден")
    
    @staticmethod
    def test_large_number():
        result = magic(277777788888899)
        expected = 11
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_large_number пройден")
    
    @staticmethod
    def test_number_1():
        result = magic(1)
        expected = 0
        assert result == expected, f"Ожидалось {expected}, получено {result}"
        print("test_number_1 пройден")

def run_magic_tests():
    print("=" * 50)
    
    TestMagic.test_single_digit()
    TestMagic.test_example_39()
    TestMagic.test_example_999()
    TestMagic.test_two_digits()
    TestMagic.test_three_digits()
    TestMagic.test_with_zero()
    TestMagic.test_another_example()
    TestMagic.test_large_number()
    TestMagic.test_number_1()
    
    print("=" * 50)
    print("Все тесты пройдены")

run_magic_tests()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Векторы должны быть одинаковой длины")

    if len(pred) == 0:
        return 0

    squared_errors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return sum(squared_errors) / len(squared_errors)


import unittest


class TestMSE(unittest.TestCase):
    def test_equal_vectors(self):
        self.assertEqual(mse([1, 2, 3], [1, 2, 3]), 0)
        self.assertEqual(mse([0, 0, 0], [0, 0, 0]), 0)

    def test_different_vectors(self):
        self.assertEqual(mse([1, 2, 3], [4, 5, 6]), 9)
        self.assertEqual(mse([1, 1], [3, 3]), 4)

    def test_negative_numbers(self):
        self.assertEqual(mse([-1, -2], [1, 2]), 10)
        self.assertEqual(mse([-5, 5], [5, -5]), 100)

    def test_decimal_numbers(self):
        self.assertAlmostEqual(mse([1.5, 2.5], [1.0, 2.0]), 0.25)
        self.assertAlmostEqual(mse([0.1, 0.2], [0.3, 0.4]), 0.04)

    def test_single_element(self):
        self.assertEqual(mse([5], [3]), 4)
        self.assertEqual(mse([10], [10]), 0)

    def test_empty_vectors(self):
        self.assertEqual(mse([], []), 0)

    def test_different_lengths(self):
        with self.assertRaises(ValueError):
            mse([1, 2], [1, 2, 3])
        with self.assertRaises(ValueError):
            mse([1], [])


if __name__ == '__main__':
    unittest.main()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    if n == 1:
        return "(1)"

    factors = {}
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1

    if n > 1:
        factors[n] = factors.get(n, 0) + 1

    result = ""
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{power})"

    return result


import unittest


class TestPrimeFactors(unittest.TestCase):
    def test_one(self):
        self.assertEqual(prime_factors(1), "(1)")

    def test_prime_numbers(self):
        self.assertEqual(prime_factors(2), "(2)")
        self.assertEqual(prime_factors(3), "(3)")
        self.assertEqual(prime_factors(17), "(17)")

    def test_powers_of_primes(self):
        self.assertEqual(prime_factors(4), "(2**2)")
        self.assertEqual(prime_factors(8), "(2**3)")
        self.assertEqual(prime_factors(9), "(3**2)")
        self.assertEqual(prime_factors(25), "(5**2)")

    def test_composite_numbers(self):
        self.assertEqual(prime_factors(6), "(2)(3)")
        self.assertEqual(prime_factors(12), "(2**2)(3)")
        self.assertEqual(prime_factors(15), "(3)(5)")

    def test_example_case(self):
        self.assertEqual(prime_factors(86240), "(2**5)(5)(7**2)(11)")

    def test_larger_numbers(self):
        self.assertEqual(prime_factors(100), "(2**2)(5**2)")
        self.assertEqual(prime_factors(12345), "(3)(5)(823)")


if __name__ == '__main__':
    unittest.main()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    if number <= 0:
        return "It is impossible"

    k = 0
    total = 0

    while total < number:
        k += 1
        total += k * k

        if total == number:
            return k

    return "It is impossible"

import unittest


class TestPyramid(unittest.TestCase):
    def test_small_numbers(self):
        self.assertEqual(pyramid(1), 1)
        self.assertEqual(pyramid(5), 2)
        self.assertEqual(pyramid(14), 3)

    def test_impossible_cases(self):
        self.assertEqual(pyramid(2), "It is impossible")
        self.assertEqual(pyramid(3), "It is impossible")
        self.assertEqual(pyramid(4), "It is impossible")
        self.assertEqual(pyramid(6), "It is impossible")
        self.assertEqual(pyramid(15), "It is impossible")

    def test_larger_numbers(self):
        self.assertEqual(pyramid(30), 4)
        self.assertEqual(pyramid(55), 5)
        self.assertEqual(pyramid(91), 6)

    def test_edge_cases(self):
        self.assertEqual(pyramid(0), "It is impossible")
        self.assertEqual(pyramid(-5), "It is impossible")

    def test_known_sequence(self):
        pyramidal_numbers = [1, 5, 14, 30, 55, 91, 140, 204, 285, 385]
        for i, num in enumerate(pyramidal_numbers, 1):
            self.assertEqual(pyramid(num), i)

if __name__ == '__main__':
    unittest.main()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number):
    digits = list(map(int, str(number)))
    n = len(digits)

    if n <= 2:
        return True

    if n % 2 == 0:
        mid_start = n // 2 - 1
        mid_end = n // 2
        left = digits[:mid_start]
        right = digits[mid_end + 1:]
    else:
        mid = n // 2
        left = digits[:mid]
        right = digits[mid + 1:]

    return sum(left) == sum(right)

import unittest


class TestIsBalanced(unittest.TestCase):
    def test_single_digit(self):
        self.assertTrue(is_balanced(5))

    def test_two_digits(self):
        self.assertTrue(is_balanced(12))
        self.assertTrue(is_balanced(99))

    def test_odd_balanced(self):
        self.assertTrue(is_balanced(121))
        self.assertTrue(is_balanced(10301))

    def test_odd_unbalanced(self):
        self.assertFalse(is_balanced(123))
        self.assertFalse(is_balanced(10234))

    def test_even_balanced(self):
        self.assertTrue(is_balanced(1221))
        self.assertTrue(is_balanced(1001))

    def test_even_unbalanced(self):
        self.assertFalse(is_balanced(1234))
        self.assertFalse(is_balanced(1002))

    def test_with_zeros(self):
        self.assertTrue(is_balanced(100001))
        self.assertFalse(is_balanced(100012))


if __name__ == '__main__':
    unittest.main()
```
